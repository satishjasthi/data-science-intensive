```python
loans = pd.concat([ldf, ldf1], axis=0)
loans
```

```python
def add_loans(csv, loans=loans):
    temp = pd.read_csv(csv, header=1)
    print "New length: " + str(len(temp)) + ", loans length: " + str(len(loans))
    if np.all(temp.columns == loans.columns):
        newloans = pd.concat([loans, temp], axis=0)
    print "New loans length: " + str(len(newloans))
    return newloans
```

```python
newloans = add_loans("data/LoanStats3c.csv")
newloans
```
